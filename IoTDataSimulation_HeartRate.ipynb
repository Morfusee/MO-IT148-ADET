```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from faker import Faker
import neurokit2 as nk
```

Main Code

---



```python
fake = Faker()
```

```python
from datetime import datetime, timedelta

start_time = datetime.strptime("00:01:00", "%H:%M:%S")
end_time = datetime.strptime("23:59:00", "%H:%M:%S")
delta = timedelta(minutes=5)

timestamps = []
current_time = start_time

while current_time <= end_time:
  timestamps.append(current_time.strftime("%H:%M:%S"))
  current_time += delta

print(timestamps)
```

```python
users = []

for n in range(user_count):
  record = {
      "id": np.random.randint(100, 999),
      "first_name": fake.first_name(),
      "last_name": fake.last_name(),
  }
  users.append(record)


# Example for Smart Healthcare Monitoring
data = []

for user in users:
  for time in timestamps:
      bpm = np.random.randint(60, 100)
      record = {
          "timestamp": time,  # Random timestamp in the last 24 hours
          "patient_id": f"PAT-{user['id']}",  # Random patient ID
          "heart_rate": bpm,  # Normal heart rate range
          "blood_pressure": {
              "systolic": np.random.randint(100, 140),
              "diastolic": np.random.randint(60, 90)
          },
          "oxygen_level": np.random.randint(95, 100),  # Oxygen saturation in normal range
          "body_temp": round(np.random.uniform(36.0, 38.0), 1),  # Body temperature in Celsius
          "ecg": nk.ecg_simulate(heart_rate=bpm, duration=1, sampling_rate=100)  # Simulate ECG signal
      }
      data.append(record)


user_df = pd.DataFrame(users)

# Save dataset
user_df.to_csv("healthcare_users.csv", index=False)
user_df.to_json("healthcare_users.json", orient="records")

# Display first few rows
user_df.head()

# Convert to DataFrame
df = pd.DataFrame(data)

# Save dataset
df.to_csv("healthcare_data.csv", index=False)
df.to_json("healthcare_data.json", orient="records")

# Display first few rows
df.head()
```
